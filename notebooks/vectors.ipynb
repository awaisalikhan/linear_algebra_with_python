```python
import matplotlib.pyplot as plt
```

```python
u = np.array([2,5])
v = np.array([3, 6])
```

```python
scatter_x, scatter_y = zip(u,v)
```

```python
scatter_x
```

```python
plt.scatter(scatter_x,scatter_y,color=["r","b"])
plt.grid()
plt.axis([0,9, 0, 9])
plt.show()
```

```python
plt.arrow()
```

```python
plt.arrow(0,0, scatter_x[0],scatter_y[1],head_width=0.2, head_length=0.3, length_includes_head=True)
```

```python
def plt_vector2d( vector2d,origin = [0, 0], **options):
    return plt.arrow(origin[0], origin[1], vector2d[0], vector2d[1],head_width=0.2, head_length=0.3, length_includes_head=True, **options)
    
```

```python
plt.grid()
```

```python
plt_vector2d(u,color='r')
plt_vector2d(v,color='g')
plt.grid()
plt.axis([0, 9, 0, 9])
plt.show()
```

```python
a = np.array([3, 5, 8])
b = np.array([2, 4, 8])
```

```python
subplot3d = plt.subplot(111, projection='3d')
x_coords, y_coords, z_coords = zip(a,b)
subplot3d.scatter(x_coords, y_coords, z_coords)
subplot3d.set_zlim3d([0, 9])
plt.show()
```

```python
def plot_vector3d(ax,vector3d, z0, **options):
    for v in vector3d:
        x, y, z = v
        ax.plot([x, x], [y, y], [z0, z], color='gray',linestyle='dotted', marker='.')
    x_coord, y_coord, z_coord, = zip(*vector3d)    
    ax.scatter(x_coord,y_coord,z_coord, **options)

        
subplot3d = plt.subplot(111, projection='3d')
subplot3d.set_zlim([0, 9])
plot_vector3d(subplot3d,[a,b],0, color=('r','b'))
plt.show()
```

```python
def norm_of_vector(vector):
    squares = [element**2  for element in vector]
    return sum(squares)**0.5
```

```python
norm_of_vector(u)
```
